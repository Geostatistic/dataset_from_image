```python
import pygeostat as gs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
```

```python
im = Image.open('image01.jpg')
```

```python
from __future__ import print_function
print(im.format, im.size, im.mode)

x_size = im.size[0]
y_size = im.size[1]
```

```python
im_data = list(im.getdata())
```

```python
xy_tuples = []
for y in reversed(range(y_size)):
    for x in range(x_size):
        xy_tuples.append([x,y])
        
for index,pixel in enumerate(xy_tuples):
    pixel.append(im_data[index])
```

```python
df = pd.DataFrame(xy_tuples, columns=['x','y','grade'])
```

```python
grade_min = 4
grade_max = 52

df['grade'] = grade_max * (df['grade'] - min(df['grade'])) / (max(df['grade']) - min(df['grade'])) + grade_min
```

```python
regular_sampling = pd.DataFrame(columns=['x','y','grade'])
```

```python
x_int = 40
y_int = 40

row_list = []

for x in range(0, x_size, x_int):
    for y in range(0, y_size, y_int):
        row = df.loc[(df['x'] == x) & (df['y'] == y)]

        regular_sampling = regular_sampling.append(row, ignore_index=True)
```

```python
print(regular_sampling)
```

```python
gs.write_gslib(regular_sampling, 'dataset01_reg_sampling.dat')
```

```python
reg_samp = gs.DataFile('dataset01_reg_sampling.dat', x='x', y='y')
```

```python
gs.locmap(reg_samp, var='grade', figsize=(10,10))
```

```python
dense_sampling = pd.DataFrame(columns=['x','y','grade'])
```

```python
cross = 5

filtro = reg_samp['grade'] > 40

for index, row in regular_sampling.iterrows():
    
    x = row['x']
    y = row['y']
    x1 = row['x'] - cross
    x2 = row['x'] + cross
    y1 = row['y'] - cross
    y2 = row['y'] + cross
    
    row1 = df.loc[(df['x'] == x) & (df['y'] == y1)]
    row2 = df.loc[(df['x'] == x) & (df['y'] == y2)]
    row3 = df.loc[(df['x'] == x1) & (df['y'] == y)]
    row4 = df.loc[(df['x'] == x2) & (df['y'] == y)]
    
    dense_sampling = dense_sampling.append(row1, ignore_index=True)
    dense_sampling = dense_sampling.append(row2, ignore_index=True)
    dense_sampling = dense_sampling.append(row3, ignore_index=True)
    dense_sampling = dense_sampling.append(row4, ignore_index=True)
```

```python
print(dense_sampling)
```

```python
gs.write_gslib(dense_sampling, 'dataset01_dense_samp.dat')
```
